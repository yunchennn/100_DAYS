```python
app_train_Con_Cash = app_train[app_train['NAME_CONTRACT_TYPE'] == 'Cash loans']
app_train_Con_Cash.head()
```

```python
app_train_Con_Cash['AMT_CREDIT'].mean()
```

```python
app_train_Con_Rev = app_train[app_train['NAME_CONTRACT_TYPE'] == 'Revolving loans']
app_train_Con_Rev.head()
```

```python
Cash_AMT_I_T = app_train_Con_Cash['AMT_INCOME_TOTAL'].mean()
Cash_AMT_C = app_train_Con_Cash['AMT_CREDIT'].mean()
Cash_AMT_A = app_train_Con_Cash['AMT_ANNUITY'].mean()
Rev_AMT_I_T = app_train_Con_Rev['AMT_INCOME_TOTAL'].mean()
Rev_AMT_C = app_train_Con_Rev['AMT_CREDIT'].mean()
Rev_AMT_A = app_train_Con_Rev['AMT_ANNUITY'].mean()
Cash = {Cash_AMT_I_T, Cash_AMT_C, Cash_AMT_A}
Rev = {Rev_AMT_I_T, Rev_AMT_C, Rev_AMT_A}
Cash
```

```python
Rev
```

```python
index=['INCOME_TOTAL', 'CREDIT', 'ANNUITY']

plt.bar(index, Cash, label='Cash', align="edge", width=0.35)
plt.bar(index, Rev, label='Revolving', align="edge", width=-0.35)
plt.legend()
plt.title("Loans Contract")
plt.xlabel("AMT")
plt.ylabel("Number")
```
